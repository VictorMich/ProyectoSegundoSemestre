```python
def factorial(num):
    fact = 1
    if num < 1:
        return "Necesita ser un numero positivo"
    elif num == 0:
        return 1
    else:
        for i in range(1,num+1):
            fact = fact * i
        return fact
```

```python
class calculadora:
    def __init__(self,a,b):
        self.a = a
        self.b = b
        
    @property
    def a(self):
        return self._a
    @property
    def b(self):
        return self._b
    
    @b.setter
    def b(self,value):
        if value == 0:
            raise ZeroDivisionError("División entre cero")
        elif type(value) != int:
            raise ValueError("No es un entero")
        self._b = value
            
    @a.setter
    def a(self,value):
        if type(value) != int:
            raise ValueError("No es un entero")
        self._a = value
            
    def suma(self):
        return self.a + self.b
    
    def resta(self):
        return self.a - self.b
    
    def multiplicacion(self):
        return self.a*self.b
    
    def division(self):
        return self.a/self.b
```
